```python
# Create dictionary, to store the single-token list of each E-SMILES
tgt_with_edit_set_train = {}
tgt_with_edit_set_test = {}

for i in range(len(tgt_data_train)):
    tgt_with_edit_set_train[i] = tgt_data_train[i]

for i in range(len(tgt_data_test)):
    tgt_with_edit_set_test[i] = tgt_data_test[i]

tgt_with_edit_set = {}
tgt_data = tgt_data_train + tgt_data_test + tgt_data_val

for i in range(len(tgt_data)):
    tgt_with_edit_set[i] = tgt_data[i]
```

```python
import re

def find_edited_indices(s):
    # Characters to find indices for
    edit_vocabs = {"^", ";", "_", "!", "~", "r", "s", "?", "α", "β", "γ", "δ", "{", "}", "&"}

    # Finding indices of special characters
    indices_dict = {char: [i for i, c in enumerate(s) if c == char] for char in edit_vocabs}
    indices_lists = [indices_dict[i] for i in indices_dict]
    indices_list = [item for sublist in indices_lists for item in sublist]
    # print(indices_list)
    
    # Finding indices of text enclosed in "<>"
    enclosed_text_indices = [[m.start(), m.end()] for m in re.finditer(r"<[^>]*>", s)]
    enclosed_indices = [list(range(i[0], i[1])) for i in enclosed_text_indices]
    # print(enclosed_indices)
    enclosed_indices_list = [item for sublist in enclosed_indices for item in sublist]
    
    return indices_list + enclosed_indices_list
```

```python
# Example of finding edit indices
s_list = tgt_with_edit_set[0]
s = "".join(s_list)
print("tgt_of_e_smiles: ", s)

edit_indices = find_edited_indices(s)
print("edit_indices: ", edit_indices)
print("edit_tokens: ", [s_list[i] for i in edit_indices])
```

```python
import joblib
DATASET = "all"
enc_outs_e_smiles = joblib.load(f"enc_outs_{DATASET}_without_rxn_e_smiles.pkl")
dec_outs_e_smiles = joblib.load(f"dec_outs_{DATASET}_without_rxn_e_smiles.pkl")
```

```python
# 将张量列表合并成一个2D张量
enc_outs_tensor_e_smiles = torch.cat([t.mean(dim=1) for t in enc_outs_e_smiles])
dec_outs_tensor_e_smiles = torch.cat([t.mean(dim=1) for t in dec_outs_e_smiles])

print("enc_outs_tensor.shape: ", enc_outs_tensor_e_smiles.shape)
print("dec_outs_tensor.shape: ", dec_outs_tensor_e_smiles.shape)
```

```python
# # Average the edit-tokens embedding

# index and label
index_of_label_list = list(tgt_with_edit_set.keys())
label_list_e_smiles = [label_list_e_smiles[i] for i in index_of_label_list]
index_of_edit_list = []
embed_of_edit_list = []

# Find the index of "specical edit tokens", then extract their embedding and mean
for i in tgt_with_edit_set.keys():
    e_smiles = "".join(tgt_with_edit_set[i])
    index_of_edit = find_edited_indices(e_smiles)
    # print("edit_indices: ", index_of_edit)
    # print("edit_tokens: ", [tgt_with_edit_set[i][j] for j in index_of_edit])
    index_of_edit_list.append(index_of_edit)
    embed_of_edit = dec_outs_e_smiles[i][:, index_of_edit, :].mean(dim=1)
    embed_of_edit_list.append(torch.squeeze(embed_of_edit))
    
print(len(index_of_label_list))
print(len(label_list_e_smiles))
print(len(index_of_edit_list))
print(len(embed_of_edit_list))

dec_outs_tensor_e_smiles = torch.stack(embed_of_edit_list)
print("dec_outs_tensor_train.shape: ", dec_outs_tensor_e_smiles.shape)
```

```python
import pandas as pd
from joblib import Parallel, delayed

df = pd.DataFrame()
df['src'] =  ["".join(i) for i in src_data]
df['tgt'] =  ["".join(i) for i in tgt_data]
df['superclasses'] = label_list_e_smiles
df['dec_out_embedding'] = dec_outs_tensor_e_smiles.cpu().tolist()
df
```

```python
import re

def find_edited_indices(s):
    # Characters to find indices for
    edit_vocabs = {"^", ";", "_", "!", "~", "r", "s", "?", "α", "β", "γ", "δ", "{", "}", "&"}

    # Finding indices of special characters
    indices_dict = {char: [i for i, c in enumerate(s) if c == char] for char in edit_vocabs}
    indices_lists = [indices_dict[i] for i in indices_dict]
    indices_list = [item for sublist in indices_lists for item in sublist]
    # print(indices_list)
    
    # Finding indices of text enclosed in "<>"
    enclosed_text_indices = [[m.start(), m.end()] for m in re.finditer(r"<[^>]*>", s)]
    enclosed_indices = [list(range(i[0], i[1])) for i in enclosed_text_indices]
    # print(enclosed_indices)
    enclosed_indices_list = [item for sublist in enclosed_indices for item in sublist]
    
    return indices_list + enclosed_indices_list
```

```python
# Example of finding edit indices
s_list = tgt_with_edit_set[0]
s = "".join(s_list)
print("tgt_of_reactseq: ", s)

edit_indices = find_edited_indices(s)
print("edit_indices: ", edit_indices)
print("edit_tokens: ", [s_list[i] for i in edit_indices])
```

```python
import joblib
dec_outs_reactseq = joblib.load(f"dec_outs_all_without_rxntype_ReactSeq.pkl")
```

### Average all tokens' embedding (ReactSeq_all)

```python
# tensor list → tensor 2D
dec_outs_tensor_reactseq = torch.cat([t.mean(dim=1) for t in dec_outs_reactseq])
print("dec_outs_tensor.shape: ", dec_outs_tensor_reactseq.shape)
```

### Average the edit-tokens' embedding (ReactSeq_MOF)

```python
# index and label
index_of_label_list = list(tgt_with_edit_set.keys())
label_list_reactseq = [label_list_reactseq[i] for i in index_of_label_list]
index_of_edit_list = []
embed_of_edit_list = []

# Find the index of "specical edit tokens", then extract their embedding and mean
for i in tgt_with_edit_set.keys():
    reactseq = "".join(tgt_with_edit_set[i])
    index_of_edit = find_edited_indices(reactseq)
    # print("edit_indices: ", index_of_edit)
    # print("edit_tokens: ", [tgt_with_edit_set[i][j] for j in index_of_edit])
    index_of_edit_list.append(index_of_edit)
    embed_of_edit = dec_outs_reactseq[i][:, index_of_edit, :].mean(dim=1)
    embed_of_edit_list.append(torch.squeeze(embed_of_edit))
    
print(len(index_of_label_list))
print(len(label_list_reactseq))
print(len(index_of_edit_list))
print(len(embed_of_edit_list))

dec_outs_tensor_reactseq = torch.stack(embed_of_edit_list)
print("dec_outs_tensor_train.shape: ", dec_outs_tensor_reactseq.shape)
```

```python
import pandas as pd
from joblib import Parallel, delayed

df = pd.DataFrame()
df['src'] =  ["".join(i) for i in src_data]
df['tgt'] =  ["".join(i) for i in tgt_data]
df['superclasses'] = label_list_reactseq
df['dec_out_embedding'] = dec_outs_tensor_reactseq.cpu().tolist()
df
```

```python
df["reactseq"] = df["src"] + ">>>" + df["tgt"] 
reactants = [cano_smiles(merge_smiles(rxn_smi)) for rxn_smi in tqdm(df["reactseq"])]
df['rxn'] = pd.Series(data=reactants) + ">>" + df["src"] 
df['class'] = df["superclasses"].apply(lambda x: x.lstrip("class_"))
```

```python
import pandas as pd


df_cat = pd.read_csv('./ELN_reaction_representation.csv')
df_cat
```

```python
value_counts = df_cat['mapped_rxn'].value_counts()
```

```python
import pandas as pd
import pandas as pd


value_counts = df_cat['mapped_rxn'].value_counts()

elements_to_find = value_counts[value_counts > 4].index

indices = df_cat[df_cat['mapped_rxn'].isin(elements_to_find)].index

print(indices)
```

```python
indices_tuples = [
    tuple(group.index) for name, group in df_cat[df_cat['mapped_rxn'].isin(elements_to_find)].groupby('mapped_rxn')
]

print(indices_tuples)
```

```python
X_test = df_cat.iloc[indices]
```

```python
X_test
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score


y = df_cat['yield']
y
```

```python
feature_col = []
for i in df_cat.columns[12:]:
    if 'mapped_rxn' == i:
        continue
    if 'ReactSeq_MEO' in i:
        continue
#     if 'ReactSeq_all' in i:
#         continue
    if 'AP3_256' in i:
        continue
    if 'Drfp' in i:
        continue
    if 'rxnfp_pretrained' in i:
        continue
    if 'rxnfp_ft_10k' in i:
        continue

    if 'amine_' in i and i != 'amine_amount':
        continue
    if 'halide_' in i:
        continue
    if 'product_' in i:
        continue
    feature_col.append(i)
```

```python
feature_col
```

```python
len(feature_col)
```

```python
X = df_cat[feature_col]
```

```python
list(indices)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import numpy as np

test_indices = list(indices)  


yield_lis = []


for seed in range(30):
    X_test = X.iloc[test_indices]
    y_test = y.iloc[test_indices]
    
    X_train = X.drop(test_indices)
    y_train = y.drop(test_indices)
    
    model = RandomForestRegressor(n_estimators=100, random_state=seed)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    
    idx_pre_yield = dict(zip(test_indices,list(y_pred)))
    idx_true_yield = dict(zip(test_indices,list(y_test)))
    pick_yield_lis = []
    mean_yield_lis = []

    for idx_tuple in indices_tuples:
        pre_yield_tuple = [idx_pre_yield[i] for i in idx_tuple]
        true_yield_tuple = [idx_true_yield[i] for i in idx_tuple]
        pick_idx = pre_yield_tuple.index(max(pre_yield_tuple))
        pick_yield_lis.append(true_yield_tuple[pick_idx])
        mean_yield_lis.append(sum(true_yield_tuple)/len(true_yield_tuple))

    
    print(seed,sum(mean_yield_lis)/len(mean_yield_lis),sum(pick_yield_lis)/len(pick_yield_lis))

    
    yield_lis.append(sum(pick_yield_lis)/len(pick_yield_lis))

print('')
print(np.mean(yield_lis))
print(np.std(yield_lis))
```
